Goal: predict a continuous number (e.g. Fare) from features (e.g. Age, Pclass). The model is just a weighted sum of inputs — same as a single neuron with no activation.

ŷ = w·x + b

ŷ is the prediction, w is a weight vector (one weight per feature), x is the input feature vector, b is the bias (intercept). This is literally np.dot(X, w) + b — the exact matrix multiply you did in cs231n.

Loss function — Mean Squared Error
L = (1/n) Σ (ŷᵢ − yᵢ)²
We square the error so positive and negative errors don't cancel out, and bigger errors get punished more. n is the number of samples.

Gradient descent — how weights get updated
∂L/∂w = (2/n) Xᵗ(ŷ − y)   |   ∂L/∂b = (2/n) Σ(ŷ − y)
w = w − α · ∂L/∂w  (α is the learning rate). We move w in the opposite direction of the gradient because the gradient points toward increasing loss — we want to decrease it.

Interview question: "Why do we use the derivative of the loss to update weights?" Answer: the gradient tells you the direction of steepest increase in loss. Moving opposite to it is the fastest way to reduce loss — this is literally what "gradient descent" means.

```python
import pandas as pd
# Load directly from URL into a DataFrame called df
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Verify it loaded correctly
print(df.shape)
```

```python
import numpy as np
class LinearRegressionScratch:
  def __init__(self,lr=0.01,n_iters=1000):
    self.lr = lr
    self.n_iters = n_iters
    self.weights = None
    self.bias = None
  def fit(self,X,y):
    n_samples,n_features=X.shape
    self.weights=np.zeros(n_features)
    self.bias=0
    for i in range(self.n_iters):
      y_pred=np.dot(X,self.weights)+self.bias
      dw=(2/n_samples)*np.dot(X.T,(y_pred-y))
      db=(2/n_samples)*np.sum(y_pred-y)
      self.weights=self.weights-self.lr*dw
      self.bias=self.bias-self.lr*db
  def predict(self,X):
    y_pred=np.dot(X,self.weights)+self.bias
    return y_pred

```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())

X = df[['Age', 'Pclass']].values
y = df['Fare'].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
my_model = LinearRegressionScratch(lr=0.1, n_iters=1000)
my_model.fit(X_scaled, y)
print("My weights:", my_model.weights, "bias:", my_model.bias)
# Sklearn's model — should give nearly identical weights
sk_model = LinearRegression()
sk_model.fit(X_scaled, y)
print("Sklearn weights:", sk_model.coef_, "bias:", sk_model.intercept_)
```
